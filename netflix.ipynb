```python
df_netflix=pd.read_csv('./datasets/netflix_data.csv')
df_netflix.head(7)
```

```python
df_netflix.isnull().sum()
```

```python
years=df_netflix.release_year.to_list()
duration=df_netflix.duration.to_list()
dict_movie={'years':years,'duration':duration}
```

```python
df_duration = pd.DataFrame(dict_movie)
df_duration.head(10)
```

```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
plt.figure(figsize=(12, 6))
plt.style.use('seaborn-v0_8-whitegrid')
sns.lineplot(x='years', y='duration', data=df_duration)
plt.title('Netflex Movies Durations')
plt.xlabel('Release Year')
plt.ylabel('Movies Durations')
plt.show()
```
